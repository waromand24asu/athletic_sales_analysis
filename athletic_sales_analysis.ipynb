```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
ath_sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
ath_sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2021 sales DataFrame

ath_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print()
ath_sales_2020_df.info()
print()
```

```python
# Check the 2021 sales data types.
print()
ath_sales_2021_df.info()
print()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

concatDF = pd.concat([ath_sales_2021_df, ath_sales_2020_df], axis=0)
SalesData_df = concatDF.reset_index()

SalesData_df
```

```python
# Check if any values are null.
null_values_df_count = SalesData_df.isna().sum()
print()
print(null_values_df_count)
print()

nValues = SalesData_df.isnull()
anyNull = nValues.any().any()

if anyNull:
    print("NULL VALUES DETECTED")
    print(nValues)
else:
    print("NO NULL VALUES DETECTED")
```

```python
# Check the data type of each column
print()
SalesData_df.info()
print()
```

```python
# Convert the "invoice_date" to a datetime datatype
SalesData_df["invoice_date"] = pd.to_datetime(SalesData_df["invoice_date"])
SalesData_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
print()
SalesData_df.info()
print()
print(f'Invoice Date data type',{SalesData_df["invoice_date"].dtype})
print()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

products_sum_df = SalesData_df.groupby(['region', 'state', 'city']).agg({'units_sold': 'sum'}).reset_index()
products_sum_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Rename the sum to "Total_Products_Sold".
products_sum_df_sorted = products_sum_df.sort_values(by=['Total_Products_Sold'], ascending=[False]).reset_index(drop=True)
products_sum_df_sorted['Total_Products_Sold'] = products_sum_df_sorted['Total_Products_Sold'].apply(lambda x: "{:,}".format(x))

# Show the top 5 results.
products_sum_df_sorted.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_products_sum_df = pd.pivot_table(SalesData_df, 
                                       values='units_sold',
                                       index=['region', 'state', 'city'],
                                       aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_products_sum_df = pivot_products_sum_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Sort df by Total_Products_Sold, descending
sorted_df = pivot_products_sum_df.sort_values(by='Total_Products_Sold', ascending=False).reset_index()
sorted_df.head(5)

# Apply formatting
sorted_df['Total_Products_Sold'] = sorted_df['Total_Products_Sold'].apply(lambda x: "{:,}".format(x))

# Show the top 5 results.
sorted_df.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

sales_sum_df = SalesData_df.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'}).sort_values(by="total_sales", ascending=False).reset_index()

# salesMade = SalesData_df.groupby(["region", "state", "city"])["total_sales"].sum().to_frame().sort_values(by="total_sales", ascending=False)
# salesMade = salesMade.rename(columns={"total_sales": "Total_Sales"})

# Rename total_sales column to Total Sales
sales_sum_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
sales_sum_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_sales_sum_df = pd.pivot_table(SalesData_df, 
                                       values='total_sales',
                                       index=['region', 'state', 'city'],
                                       aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_sales_sum_df = pivot_sales_sum_df.rename(columns={'total_sales': 'Total Sales'})

# Sort df by Total_Products_Sold, descending
sorted_sales_df = pivot_sales_sum_df.sort_values(by='Total Sales', ascending=False).reset_index()

# Show the top 5 results.
sorted_sales_df.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sum_df = SalesData_df.groupby(['retailer','region', 'state', 'city']).agg({'total_sales': 'sum'}).reset_index()

# Rename the "total_sales" column to "Total Sales"
retailer_sum_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Sort df by Total Sales, descending
retailer_sum_df_sorted = retailer_sum_df.sort_values(by=['Total_Sales'], ascending=[False]).reset_index(drop=True)

# Show the top 5 results.
retailer_sum_df_sorted.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_retailer_sales_sum_df = pd.pivot_table(SalesData_df, 
                                       values='total_sales',
                                       index=['retailer', 'region', 'state', 'city'],
                                       aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_retailer_sales_sum_df = pivot_retailer_sales_sum_df.rename(columns={'total_sales': 'Total_Sales'})

# Sort df by Total_Products_Sold, descending
sorted_pivot_retailer_sales_df = pivot_retailer_sales_sum_df.sort_values(by='Total_Sales', ascending=False).reset_index()

# Show the top 5 results.
sorted_pivot_retailer_sales_df.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
unique_values = pd.unique(SalesData_df['product'])

womens_ath_footware_df = SalesData_df.loc[SalesData_df['product'] == "Women's Athletic Footwear"]

# Print table
womens_ath_footware_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_ath_retailer_sum_df = womens_ath_footware_df.groupby(['retailer','region', 'state', 'city']).agg({'units_sold': 'sum'}).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_ath_retailer_sum_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Sort df by Total Sales, descending
womens_ath_retailer_sum_df_sorted = womens_ath_retailer_sum_df.sort_values(by=['Womens_Footwear_Units_Sold'], ascending=[False]).reset_index(drop=True)

# Show the top 5 results.
womens_ath_retailer_sum_df_sorted.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_womens_ath_retailer_product_sum_df = pd.pivot_table(womens_ath_footware_df, 
                                       values='units_sold',
                                       index=['retailer', 'region', 'state', 'city'],
                                       aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_womens_ath_retailer_product_sum_df = pivot_womens_ath_retailer_product_sum_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Sort df by Total_Products_Sold, descending
sorted_womens_ath_retailer_product_sum_df = pivot_womens_ath_retailer_product_sum_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).reset_index()

# Show the top 5 results.
sorted_womens_ath_retailer_product_sum_df.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
women_footwear_sales_pivot_date = pd.pivot_table(womens_ath_footware_df, values='total_sales', index='invoice_date', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
women_footwear_sales_pivot_date.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
print("\nWomen's Athletic Footwear Sales by Day:")
women_footwear_sales_pivot_date.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
dailySales_df = women_footwear_sales_pivot_date.resample("D").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
dailySales_df = dailySales_df.sort_values(by="Total Sales", ascending=False)

# Sort the resampled pivot table in descending order on "Total Sales".
dailySales_df.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales_df = women_footwear_sales_pivot_date.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
print("\nTop 10 Weeks with Most Women's Athletic Footwear Sales:")
print(weekly_sales_df.sort_values(by='Total Sales', ascending=False).head(10))
```
